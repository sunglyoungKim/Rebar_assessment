```python
import random
import numpy as np
import string
```

```python
function_list = [f'generate_type_{f}'for f in range(10)] # currently you have to choose the last number of generate_type

def generate_type_0():
    '''
    Uppercase-Digits or with parenthis 
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.ascii_uppercase, k=prefix_n))
    suffix= ''.join(random.choices(string.digits, k=suffix_n))
    if np.random.uniform() > 0.5:
        return prefix + '-' + suffix
    else:
        return '(' + prefix + '-' + suffix + ')'


def generate_type_1():
    '''
    Uppercase Digits
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.ascii_uppercase, k=prefix_n))
    suffix= ''.join(random.choices(string.digits, k=suffix_n))
    if np.random.uniform() > 0.5:
        return prefix + ' ' + suffix
    else:
        return '(' + prefix + ' ' + suffix + ')'


def generate_type_2():
    '''
    lowercase-Digits
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.ascii_lowercase, k=prefix_n))
    suffix= ''.join(random.choices(string.digits, k=suffix_n))
    if np.random.uniform() > 0.5:
        return prefix + '-' + suffix
    else:
        return '(' + prefix + '-' + suffix + ')'

def generate_type_3():
    '''
    lowercase Digits
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.ascii_lowercase, k=prefix_n))
    suffix= ''.join(random.choices(string.digits, k=suffix_n))
    if np.random.uniform() > 0.5:
        return prefix + ' ' + suffix
    else:
        return '(' + prefix + ' ' + suffix + ')'


def generate_type_4():
    '''
    Digits-Uppercase
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.digits, k=prefix_n))
    suffix= ''.join(random.choices(string.ascii_uppercase, k=suffix_n))
    
    if np.random.uniform() > 0.5:
        return prefix + '-' + suffix
    else:
        return '(' + prefix + '-' + suffix + ')'


def generate_type_5():
    '''
    Digits-Uppercase
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.digits, k=prefix_n))
    suffix= ''.join(random.choices(string.ascii_uppercase, k=suffix_n))
    
    if np.random.uniform() > 0.5:
        return prefix + ' ' + suffix
    else:
        return '(' + prefix + ' ' + suffix + ')'

def generate_type_6():
    '''
    Digits-lowercase
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.digits, k=prefix_n))
    suffix= ''.join(random.choices(string.ascii_lowercase, k=suffix_n))

    if np.random.uniform() > 0.5:
        return prefix + '-' + suffix
    else:
        return '(' + prefix + '-' + suffix + ')'

def generate_type_7():
    '''
    Digits lowercase
    '''
    prefix_n, suffix_n = random.choices(np.arange(1, 10), k=2)
    prefix = ''.join(random.choices(string.digits, k=prefix_n))
    suffix= ''.join(random.choices(string.ascii_lowercase, k=suffix_n))
    if np.random.uniform() > 0.5:
        return prefix + ' ' + suffix
    else:
        return '(' + prefix + ' ' + suffix + ')'    
    
def generate_type_8():
    '''
    HXM 18"x20"
    '''
    
    prefix_n = random.choice(np.arange(1, 10))
    suffix_n = random.choice(np.arange(1, 3))
    
    prefix = ''.join(random.choices(string.ascii_uppercase, k=prefix_n))
    suffix_0= ''.join(random.choices(string.digits, k=suffix_n))+'"'
    suffix_1= ''.join(random.choices(string.digits[1:], k=suffix_n))+'"'

    if np.random.uniform() > 0.5:
        return prefix + ' ' + suffix_0 + 'x' + suffix_1
    else:
        return '(' + prefix + ' ' + suffix_0 + 'x' + suffix_1+ ')'

def generate_type_9():
    '''
    18"x20" HXM
    '''
    prefix_n = random.choice(np.arange(1, 10))
    suffix_n = random.choice(np.arange(1, 3))
    
    prefix = ''.join(random.choices(string.ascii_uppercase, k=prefix_n))
    suffix_0= ''.join(random.choices(string.digits, k=suffix_n))+'"'
    suffix_1= ''.join(random.choices(string.digits[1:], k=suffix_n))+'"'
    
    if np.random.uniform() > 0.5:
        return suffix_0 + 'x' + suffix_1 + ' ' + prefix
    else:
        return '(' + suffix_0 + 'x' + suffix_1 + ' ' + prefix + ')'
        

def generate_enum(n_gen):
    gen_texts = []
    for i in range(n_gen):
        fun = np.random.choice(function_list)
        match fun:
            case 'generate_type_0':
                gen_text= generate_type_0()
            case 'generate_type_1':
                gen_text= generate_type_1()
            case 'generate_type_2':
                gen_text= generate_type_2()
            case 'generate_type_3':
                gen_text= generate_type_3()
            case 'generate_type_4':
                gen_text= generate_type_4()
            case 'generate_type_5':
                gen_text= generate_type_5()
            case 'generate_type_6':
                gen_text= generate_type_6()                
            case 'generate_type_7':
                gen_text= generate_type_7()
            case 'generate_type_8':
                gen_text= generate_type_8()
            case 'generate_type_9':
                gen_text= generate_type_9()
                
        gen_texts.append(gen_text)

    # return gen_texts
    
    with open('/home/sk/repo/Rebar_assessment/rebar_gen/text/rebar_data.txt', 'w') as outfile: # Please put your dest_path
      outfile.write('\n'.join(str(i) for i in gen_texts))
```
